```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Step 1: Define the data
print("Step 1: Defining sample data")
X = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10]).reshape(-1, 1)  # Hours studied
Y = np.array([0, 0, 0, 0, 1, 0, 1, 1, 1])  # Pass (1) or Fail (0)
print(f"Hours studied: {X.flatten()}")
print(f"Exam result (Pass=1 / Fail=0): {Y}")

# Step 2: Fit the logistic regression model
print("\nStep 2: Fitting Logistic Regression Model")
model = LogisticRegression()
model.fit(X, Y)

# Get coefficients
beta_0 = model.intercept_[0]
beta_1 = model.coef_[0][0]
print(f"Estimated equation: logit(p) = {beta_0:.4f} + {beta_1:.4f} * X")

# Step 3: Interpret the coefficient
print("\nStep 3: Interpretation")
print(f"For every additional hour studied, the log-odds of passing increase by {beta_1:.4f}")

# Step 4: Predict probability for 5 hours (example from file)
hours = 5
logit_val = beta_0 + beta_1 * hours
p = 1 / (1 + np.exp(-logit_val))
print(f"\nStep 4: Prediction for a student who studies {hours} hours:")
print(f"logit(p) = {logit_val:.4f}")
print(f"Probability of passing = {p:.4f} or {p*100:.2f}%")

# Step 5: Visualize the logistic regression curve (like in the file)
print("\nStep 5: Visualizing Logistic Curve")
hours_range = np.linspace(0, 12, 200).reshape(-1, 1)
probabilities = model.predict_proba(hours_range)[:, 1]

plt.figure(figsize=(8, 6))
plt.plot(hours_range, probabilities, color='blue', linewidth=2)
plt.scatter(X, Y, color='orange', edgecolor='black', s=80)
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression: Probability of Passing vs Hours Studied')
plt.grid(True, alpha=0.3)
plt.show()
```
